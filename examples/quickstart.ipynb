# matmercator quickstart

A minimal end-to-end run on the tiny committed fixture (`tests/fixtures/mp_20/`, 60 structures), so it works without downloading the full dataset.

> The fixture is intentionally tiny, so the metrics below are weak — this notebook demonstrates the API, not a scientific result. For the real benchmark use `matmercator run` (or point `data_root` at `data/`).

```python
from matmercator.config import PipelineConfig
from matmercator.pipeline import run

cfg = PipelineConfig(
    data_root="../tests/fixtures",
    dataset="mp_20",
    splits=("train", "val"),
    frame_set_size=30,
    gtm_k=6,
    gtm_niter=120,
    n_permutations=100,
    output_dir="../results/_notebook",
)
report = run(cfg)
print("structures:", report["n_structures"], "| frame set:", report["n_frame_set"])
for name, m in report["metrics"].items():
    print(f"{name}: obs={m['observed']:.3f}  z={m['z']:.1f}  p={m['p_value']:.3f}")
```
